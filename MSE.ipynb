```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
x_values = [i*14 for i in range(15)]

# Generate some dummy data for 'multi' and 'stat' since it's not provided
success = [1]*15

SSIM = [1]*15
MSE = [0]*15

marker_size = 11
# Plot lines with the new styles and markers
ax.plot(x_values, success, label='success_rate', color='blue', linestyle='--', marker='o', linewidth=2,markersize=marker_size)
ax.plot(x_values, SSIM, label='SSIM', color='red', linestyle='--', marker='x', linewidth=2,markersize=marker_size)
ax.plot(x_values, MSE, label='MSE', color='g', linestyle='--', marker='*', linewidth=2,markersize=marker_size)

# Add grid lines
ax.grid(True, which='both', linestyle='-', linewidth=0.5)
for spine in ax.spines.values():
    spine.set_linewidth(2)
# Set axis labels
ax.set_xlabel('Epoch', fontsize=12)
#ax.set_ylabel('Running time /s', fontsize=12)

# Set the legend to match the style in the second plot
#legend = ax.legend(frameon=False, loc='upper left', bbox_to_anchor=(1,1))
legend = ax.legend(frameon=True, loc='right', fontsize=12)
# Set the title if necessary
#ax.set_title('Running time as a function of m')

# Show the plot with tight layout to ensure the legend is placed correctly
plt.tight_layout()

# 保存为PDF格式
plt.savefig('./figure/running_time.pdf', bbox_inches='tight')

# 保存为PNG格式
plt.savefig('./figure/running_time.png', bbox_inches='tight')

plt.show()
```
