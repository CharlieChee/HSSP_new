```python
import numpy as np
import matplotlib.pyplot as plt



# 
R = np.load('/Users/jichanglong/Desktop/hssp_new/expdata/true_fl/cifar10/truefl/images_epoch_4_batch_0.npy')


# [-1, 1][0,1]
R = (R + 1.0) / 2.0

# [0,255]
R = (R * 255.0).astype(np.uint8)

# (batch_size, height, width, channels)
R = np.transpose(R, (0, 2, 3, 1))

np.random.shuffle(R)
#R=np.random.shuffle(R)
#print(R[0,:])
n=10
#idx=matrix()

import matplotlib.pyplot as plt
num_col = 10
num_row = int(n/num_col)
num_samples = num_row*num_col
shape= (32,32,3)
fig, axes = plt.subplots(num_row, num_col, figsize=(1.*num_col,1.*num_row))
print('\n Original input images:')
for i, ax in enumerate(axes.flat):
      ax.set_axis_off()
      ax.imshow(R[i].reshape(shape), cmap='gray') 
plt.savefig('./figure/true/cifar10_stat.pdf', bbox_inches='tight')

# PNG
plt.savefig('./figure/true/cifar10_stat.png', bbox_inches='tight')
```
